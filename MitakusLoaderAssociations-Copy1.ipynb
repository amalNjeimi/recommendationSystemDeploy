```python
for f in files:
    p = pd.read_csv('C:/Users/Njeimi Amal/Desktop/pi/AssociationRulesRestSpring/a/'+f,sep=";")
    p.Antecedent=p.Antecedent.str.replace("'", " ")
    p.Consequent=p.Consequent.str.replace("'", " ")
    p.Consequent=p.Consequent.str.replace(",,", " ")
    p.Consequent=p.Consequent.str.replace(" ,", "|")
    p.Antecedent=p.Antecedent.str.replace(" ,", "|")
    cols = p.columns.values
    val = f.replace('.csv', '').lower()
    cursor = db.cursor()
    t1="DROP TABLE IF EXISTS `"+val+"`"
    cursor.execute(t1)
    table='create table '+val+'(id int primary key AUTO_INCREMENT, Antecedent TEXT(500), Consequent TEXT (50))'
    cursor.execute(table)
    for c in cols:
        for row in p.itertuples():
            
            query = "INSERT INTO `"+ val + "` (Antecedent, Consequent) VALUES(%s, %s)"
                
            values = (row.Antecedent, row.Consequent)
            try:
                cursor.execute(query, values)
            except:
                print("This is an error message!") 
                
            print(cursor.lastrowid, "record inserted")
    db.commit()
```
